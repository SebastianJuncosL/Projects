```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
mu_0 = (4*np.pi)*1e-6
```

```python
r0 = [0,0]
m = [0,0.1]
```

```python
N = 21

gridMax = 1
gridMin = -1

x = np.linspace(gridMin, gridMax, N)
y = np.linspace(gridMin, gridMax, N)

X, Y = np.meshgrid(x, y)
```

```python
XNew = X - r0[0]
YNew = Y - r0[1]
```

```python
cons = mu_0/(4*np.pi)
```

```python
mag_XY = np.sqrt(XNew**2 + YNew**2)

X_unit = XNew/mag_XY
Y_unit = YNew/mag_XY
```

```python
r = np.sqrt(XNew**2 + YNew**2)
```

```python
m_dot_r = (m[0]*X_unit)+(m[1]*Y_unit)
```

```python
B_x = cons*((3 * X_unit * m_dot_r - m[0]) / r**3)
B_y = cons*((3 * Y_unit * m_dot_r - m[1]) / r**3)

mag_B = np.sqrt(B_x**2 + B_y**2)

B_x_unit = B_x / mag_B
B_y_unit = B_y / mag_B
```

```python
plt.figure(figsize=(16,8))

plt.subplot(1,2,1)
plt.scatter(r0[0],r0[1], c='green', s=500)
plt.quiver(x,y,B_x/mag_B,B_y/mag_B)
plt.title('Vectores Unitarios del Campo Magnético')

plt.subplot(1,2,2)
plt.streamplot(X,Y,B_x,B_y)
plt.scatter(r0[0],r0[1], c='green', s=500)
plt.title('Lineas de trayectoria del campo Magnético')
```
